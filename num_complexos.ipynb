```python
import matplotlib.pyplot as plt

# Número complexo
z = 3 + 4j

# Plotando o número complexo no plano
plt.figure(figsize=(6, 6))
plt.plot(z.real, z.imag, 'ro')  # 'ro' significa ponto vermelho
plt.text(z.real + 0.1, z.imag, f'z = {z}', fontsize=12)

# Configuração dos eixos
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.title('Representação de um Número Complexo no Plano')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Lista de números complexos
numeros_complexos = [3 + 4j, 1 - 2j, -2 + 3j, -4 - 1j, 2 + 2j]

# Plotando os números complexos
plt.figure(figsize=(6, 6))
for z in numeros_complexos:
    plt.plot(z.real, z.imag, 'o', label=f'z = {z}')

# Adicionando rótulos e eixos
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.title('Vários Números Complexos no Plano')

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Número complexo
z = 3 + 4j
conjugado = z.conjugate()

# Plotando os números
plt.figure(figsize=(6, 6))
plt.plot(z.real, z.imag, 'ro', label=f'z = {z}')  # Ponto para o número complexo
plt.plot(conjugado.real, conjugado.imag, 'bo', label=f'Conjugado de z = {conjugado}')  # Ponto para o conjugado

# Conectando o número complexo e seu conjugado com uma linha
plt.plot([z.real, conjugado.real], [z.imag, conjugado.imag], 'k--')

# Adicionando rótulos e eixos
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.title('Número Complexo e Seu Conjugado')

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Número complexo
z = 3 + 4j

# Módulo e argumento
modulo = abs(z)
argumento = np.angle(z)

# Plotando o número complexo
plt.figure(figsize=(6, 6))
plt.plot(z.real, z.imag, 'ro', label=f'z = {z}')

# Representando o módulo
plt.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color='g', label=f'Módulo = {modulo:.2f}')

# Adicionando rótulos e eixos
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.title('Módulo e Argumento de um Número Complexo')

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```
