```python
# Gráfico de um Número complexo
z = 3 + 4j

# Plotando o número complexo no plano
plt.figure(figsize=(6, 6))
plt.plot(z.real, z.imag, 'ro')  # 'ro' significa ponto vermelho

plt.text(z.real + 0.1, z.imag, f'z = {z}', fontsize=12)

# Configuração dos eixos
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

plt.xlim(0, 5)
plt.ylim(0, 5)

plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.title('Representação de um Número Complexo no Plano')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

## Múltiplos números

```python
# Lista de números complexos
numeros_complexos = [3 + 4j, 1 - 2j, -2 + 3j, -4 - 1j, 2 + 2j]

# Plotando os números complexos
plt.figure(figsize=(6, 6))

for z in numeros_complexos:
    plt.plot(z.real, z.imag, 'o', label=f'z = {z}')

# Adicionando rótulos e eixos
plt.axhline(0, color='red',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.title('Vários Números Complexos no Plano')

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

## Plot do Complexo Conjugado

```python
# Número complexo
z = 3 + 4j

conjugado = z.conjugate()

# Plotando os números
plt.figure(figsize=(6, 6))

plt.plot(z.real, z.imag, 'ro', label=f'z = {z}')  # Ponto para o número complexo
plt.plot(conjugado.real, conjugado.imag, 'bo', label=f'Conjugado de z = {conjugado}')  # Ponto para o conjugado

# Conectando o número complexo e seu conjugado com uma linha
plt.plot([z.real, conjugado.real], [z.imag, conjugado.imag], 'k--')

# Adicionando rótulos e eixos
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.title('Número Complexo e Seu Conjugado')

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

# Biblioteca CMATH

```python
import cmath
```

```python
# Definindo um número complexo
z = 3 + 4j
```

```python
cmath.phase(z)
```

```python
# Convertendo para coordenadas polares (módulo e argumento)
modulo, argumento = cmath.polar(z)

print(f'Módulo: {modulo}')
print(f'Argumento (em radianos): {argumento}')
```

## Convertendo Polar para Retangular

```python
# Definindo módulo e argumento
modulo = 5
#argumento = cmath.pi / 4  # 45 graus em radianos

argumento = 0.9272952180016122

# Convertendo para forma retangular
z_retangular = cmath.rect(modulo, argumento)

print(f'Número complexo na forma retangular: {z_retangular}')
```

## Operações com Números Complexos

```python
z1 = 3 + 4j
z2 = 1 + 2j

# Convertendo ambos para polar
modulo1, arg1 = cmath.polar(z1)
modulo2, arg2 = cmath.polar(z2)

# Agora, somando em coordenadas polares (não trivial, mas possível)
# Converter de volta para retangular antes de somar
z1_ret = cmath.rect(modulo1, arg1)
z2_ret = cmath.rect(modulo2, arg2)

soma = z1_ret + z2_ret

print(f'Soma: {soma}')
```

```python
# Definindo dois números complexos
z1 = 3 + 4j
z2 = 1 + 2j

# Convertendo para coordenadas polares
mod1, arg1 = cmath.polar(z1)
mod2, arg2 = cmath.polar(z2)

# Realizando a divisão (dividir os módulos e subtrair os argumentos)
mod_div = mod1 / mod2
arg_div = arg1 - arg2

# Convertendo o resultado de volta para forma retangular
z_div = cmath.rect(mod_div, arg_div)

print(f'Divisão em coordenadas polares: {z_div}')
print(f'Módulo do resultado: {mod_div}')
print(f'Argumento do resultado (em radianos): {arg_div}')
```
